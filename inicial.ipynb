# Projeto Python IA: Inteligência Artificial e Previsões

### Case: Score de Crédito dos Clientes

Você foi contratado por um banco para conseguir definir o score de crédito dos clientes. Você precisa analisar todos os clientes do banco e, com base nessa análise, criar um modelo que consiga ler as informações do cliente e dizer automaticamente o score de crédito dele: Ruim, Ok, Bom

Arquivos da aula: https://drive.google.com/drive/folders/1FbDqVq4XLvU85VBlVIMJ73p9oOu6u2-J?usp=drive_link

```python
# Passo 0: Entender o que o cliente irá precisar de acordo com o problema que precisa de solução

import pandas as pd

# Passo 1: Importar a base de dados
tabela = pd.read_csv("clientes.csv")
display(tabela)
```

```python
# Passo 2: Tratamento de dados
    # Arrumar erros no banco de dados
display(tabela.info())
```

```python
# Passo 3: Tratamento de dados para IA (transformar caracters em numeros para leitura)

# From = Acessa sklearn(biblioteca) tem o pacote (preprocessing, um conjunto de ação) fazendo com que eu chegue no caminho LabelEncoder que ira transformar dados em codigos(numeros)
from sklearn.preprocessing import LabelEncoder

codificador = LabelEncoder()

# Coluna profissao (LabelEncoder + fit_transform) irá pegar os dados em caracteres e transformar em numeros para IA entender
tabela["profissao"] = codificador.fit_transform(tabela["profissao"])


# Coluna mix_credito
tabela["mix_credito"] = codificador.fit_transform(tabela["mix_credito"])

# Coluna comportamento_pagamento
tabela["comportamento_pagamento"] = codificador.fit_transform(tabela["comportamento_pagamento"])

display(tabela)
```

```python
# Separação de dados para IA conseguir dar o score final
x = tabela.drop(columns=["id_cliente", "score_credito"])
y = tabela["score_credito"]

# Função que irá fazer a separação da tabela (base de dados) em elementos para IA entender e ser testada
from sklearn.model_selection import train_test_split

# train_teste_split = pega a tabela x que sao dados que a IA precisa usar para prever e o y é a previsão feita, e essas tabelas sao separadas em treino e teste para o treinamento e teste da IA
x_treino, x_teste, y_treino, y_teste = train_test_split(x,y)

# Passo 4: Criar um modelo que consiga ler o score de credito

# Processo de criação de uma IA
# Importar IA
        # Arvore de Decisão faz perguntas para base de dados
from sklearn.ensemble import RandomForestClassifier

# Modelo vizinhos proximos KNN
from sklearn.neighbors import KNeighborsClassifier

# Criar IA
modelo_arvore_decisao = RandomForestClassifier()
modelo_knn = KNeighborsClassifier()

# Treinando IA
modelo_arvore_decisao.fit(x_treino, y_treino)
modelo_knn.fit(x_treino, y_treino)

# Testar IA (Prever)
previsao_arvore_decisao = modelo_arvore_decisao.predict(x_teste)
previsao_modelo_knn = modelo_knn.predict(x_teste)

# Precisão de Acerto da IA
from sklearn.metrics import accuracy_score

display(accuracy_score(y_teste, previsao_arvore_decisao))
display(accuracy_score(y_teste, previsao_modelo_knn))
```

```python
# Passo 5: Previsao de dados

tabela_cliente_novo = pd.read_csv("novos_clientes.csv")

tabela_cliente_novo["profissao"] = codificador.fit_transform(tabela_cliente_novo["profissao"])

tabela_cliente_novo["mix_credito"] = codificador.fit_transform(tabela_cliente_novo["mix_credito"])

tabela_cliente_novo["comportamento_pagamento"] = codificador.fit_transform(tabela_cliente_novo["comportamento_pagamento"])

previsao_score = modelo_arvore_decisao.predict(tabela_cliente_novo)

display(tabela_cliente_novo)
display(previsao_score)
```
